```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/Budget-Sample .csv')
df['base_salary'].hist(bins = 50, figsize = (11, 8), color = 'green')
plt.xlabel('Base Salary')
plt.ylabel('Number of People who have it')
```

# There are tons of base salaries that are less than 100k and very little salaries above 150k and the most common salary is around 80k

```python
gb = df.groupby('agency_name')
gb['base_salary'].mean().sort_values().plot(kind = 'barh', figsize = (9, 60))
```

# Based on these salaries I would love to work for NYC Fire Pension Fund

```python
df = pd.read_csv('data/BX-Police.csv')
gb = df.groupby('fiscal_year')
gb['base_salary'].mean().plot(kind = 'line', figsize = (11, 8))
plt.ylabel('Mean Base Salary')
```

# This trend line shows that generally every fiscal year the Mean Base Salary goes up but it did take a dip in 2019

```python
df = pd.read_csv('data/Iris.csv')
colors = {"Iris-setosa": "red", "Iris-versicolor": "green", "Iris-virginica": "blue"}
fig, ax = plt.subplots()

for species in df["class"].unique():
    subset = df[df["class"] == species]
    ax.scatter(x=subset["petal_length"], y=subset["petal_width"],
               alpha=0.5, label = species, c = colors[species])

ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
ax.legend()

plt.figure(figsize = (11,21))
plt.show()
```

# The setosa has a small petal size, the versicolor have a medium petal size, and the virginica have the biggest petal size

```python
df = pd.read_csv('data/Sales.csv')

features = df.columns[1:]
df.plot(kind = 'line', x = 'month_number', y = features, figsize = (11,8))
plt.ylabel('Number of Sales')
plt.xlabel('Month of Sale')
```

# This graph shows me that the items sold are very constant in their sales over the months. 

# With bathing soap and toothpaste constantly selling high numbers and growing in sales, so I would push these products more.

# While the facecream, moisturzer and shampoo are not growing in sales as much so I would not push these products as much

```python
total_sales = df[features].sum().sum()
items = df[features].sum()

percentage = items/total_sales * 100

percentage.plot.pie(y = features,
                    startangle = 90,
                    autopct = '%1.1f%%',
                    legend = True,
                    ylabel = '',
                    title = 'Sales Percentage',
                    labeldistance = None,
                    figsize = (13,11)
                   )
```
